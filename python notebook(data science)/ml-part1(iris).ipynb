### Predict sepal length

```python
iris_df['species']=iris_df['species'].map({'setosa':0,'vesicolor':1,'virginica':2})
iris_df
```

```python
X=iris_df.drop(labels='sepal length (cm)',axis=1)
Y=iris_df['sepal length (cm)']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
mod=model.predict(X_test)
```

```python
mod
```

```python
b = pd.concat([X_test, Y_test], axis=1)
b.head()
```

```python
b['pred_sepal_length'] = mod
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('Mean Absolute Error:', mean_absolute_error(Y_test, mod))
print('Mean Squared Error:', mean_squared_error(Y_test, mod))
print('Mean Root Squared Error:', np.sqrt(mean_squared_error(Y_test, mod)))
```
